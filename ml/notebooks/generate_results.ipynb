# Generate model_results.xlsx by combining results from all three diseases.

Run this after training all models.

---

```python
import pandas as pd
import os

results = []

for disease in ['heart', 'diabetes', 'kidney']:
    path = f'../results/{disease}_results.csv'
    if os.path.exists(path):
        df = pd.read_csv(path)
        results.append(df)

if results:
    combined = pd.concat(results, ignore_index=True)
    combined = combined[['Disease', 'Model', 'Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC']]
    combined.to_excel('../results/model_results.xlsx', index=False)
    print("model_results.xlsx generated successfully!")
    print("\n", combined.to_string(index=False))
else:
    print("No results found. Please run training scripts first.")
```
